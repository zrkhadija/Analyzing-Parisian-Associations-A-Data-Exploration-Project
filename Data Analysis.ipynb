```python
import pandas as pd
import requests

# Replace 'your_api_endpoint' with the actual API endpoint you want to use
api_endpoint = 'https://opendata.paris.fr/api/explore/v2.1/catalog/datasets/liste_des_associations_parisiennes/records?limit=20'

# Make a GET request to the API
response = requests.get(api_endpoint)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON data from the response
    data = response.json()

    # Extract relevant data from the JSON structure
    results = data.get('results', [])

    if results:
        # Convert the JSON data to a DataFrame
        df = pd.DataFrame(results)

        # Now you can work with the DataFrame (e.g., print it)
        print(df)
    else:
        print("No results found in the JSON data.")
else:
    # Print an error message if the request was not successful
    print(f"Error: {response.status_code}")
```

### Bar Chart for Sector of Activity:
Visualize the distribution of organizations based on their primary sector of activity.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.countplot(x='sa_secteur_d_activit_1', data=df)
plt.title('Distribution of Organizations by Sector of Activity')
plt.xlabel('Sector of Activity')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Geographical Distribution using Pie Chart:
Show the distribution of organizations across different geographical sectors.

```python
plt.figure(figsize=(8, 8))
df['sg_secteur_gographique'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Geographical Distribution of Organizations')
plt.ylabel('')
plt.show()
```

```python
# Create a DataFrame with the counts for each combination of sector and geographic place
df_counts = df.groupby(['sg_secteur_gographique', 'sa_secteur_d_activit_1']).size().unstack(fill_value=0)

# Ensure the sectors are ordered for plotting
ordered_sectors = df['sa_secteur_d_activit_1'].value_counts().index
df_counts = df_counts.reindex(columns=ordered_sectors)

# Calculate the percentage of sectors for every place
sector_percentage_by_place = df_counts.div(df_counts.sum(axis=1), axis=0) * 100

# Plotting the stacked bar chart
plt.figure(figsize=(16, 8))
sector_percentage_by_place.plot(kind='bar', stacked=True, colormap='viridis', figsize=(14, 8))
plt.title('Primary Sector of Activity by Geographic Place')
plt.xlabel('Geographic Place')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Primary Sector of Activity', bbox_to_anchor=(1, 1))

# Adding text labels for sector percentages
for i, (place, percentages) in enumerate(sector_percentage_by_place.iterrows()):
    for sector, percentage in percentages.items():
        plt.text(i, percentages.head(ordered_sectors.tolist().index(sector)).sum() + percentage / 2, f'{percentage:.1f}%', ha='center', va='center')

plt.show()
```

```python
from itertools import combinations
from collections import Counter
import networkx as nx
import matplotlib.pyplot as plt

# Selecting the activity sector columns
activity_sectors = df[['sa_secteur_d_activit_1', 'sa_secteur_d_activit_2', 'sa_secteur_d_activit_3']].dropna()

# Create a list of all unique pairs of activity sectors
all_pairs = []
for _, row in activity_sectors.iterrows():
    sectors = row.dropna().tolist()
    all_pairs.extend(combinations(sectors, 2))

# Count the occurrences of each pair
pair_counts = Counter(all_pairs)

# Calculate the total number of organizations
total_organizations = len(activity_sectors)

# Calculate the probability of each pair
pair_probabilities = {pair: count / total_organizations for pair, count in pair_counts.items()}

# Extract pairs and probabilities for visualization
pairs = list(pair_probabilities.keys())
probabilities = list(pair_probabilities.values())

# Create a graph for visualization
G = nx.Graph()

# Add nodes and edges to the graph
for pair, probability in pair_probabilities.items():
    G.add_edge(pair[0], pair[1], weight=probability)

# Draw the graph with larger figure size and font size
plt.figure(figsize=(14, 10))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=1200, node_color='skyblue', font_size=12)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(edge[0], edge[1]): f'{prob:.2%}' for edge, prob in pair_probabilities.items()}, font_size=10)

plt.title('Probability of Co-occurrence of Activity Sectors')
plt.show()
```
